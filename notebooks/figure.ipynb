```python
%matplotlib widget

import sympy as sp
import matplotlib.pyplot as plt

from mechpy.core.symbolic.coord import (
    SymbolicCartesianCoordSystem,
    SymbolicCylindricalCoordSystem,
)

from mechpy.core.symbolic.field import (
    SymbolicScalarField,
    SymbolicVectorField,
    SymbolicTensorField,
)

from mechpy.utils.figure import (SymbolicScalarFieldFigure, SymbolicVectorFieldFigure)

plt.ioff()
```

```python
n, m = sp.symbols("n m")
field_params = {
    n: (0.1, 0.2, 0.5, 1, 2, 3, 5),
    m: (0.1, 0.2, 0.5, 1, 2, 3, 5),
}
data = sp.Array([1*n, -2*m, 0])
linear_scalar_field = SymbolicScalarField.create_linear(data, field_params=field_params)
display(linear_scalar_field)
```

```python
figure = SymbolicScalarFieldFigure(linear_scalar_field)
figure.plot()
```
